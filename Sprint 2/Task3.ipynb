```
public class Employee
{
    public string Name;
    public int Age;
    public int Salary;
    public string Department; 
    public int EmployeeId; 
    public DateTime HireDate;

    public Employee(string name, int age, int salary, string department, int employeeId, DateTime hireDate)
    {
        Name = name;
        Age = age;
        Salary = salary;
        Department = department;
        EmployeeId = employeeId;
        HireDate = hireDate;
    }

    public virtual string GetInfo()
    {
        return $"Имя: {Name}, Возраст: {Age}, Зарплата: {Salary} рублей, Отдел: {Department}";
    }

    public virtual void Work()
    {
        Console.WriteLine($"Сотрудник {Name} работает");
    }

    public virtual void TakeVacation()
    {
        Console.WriteLine($"Сотрудник {Name} взял отпуск");
    }

    public virtual void Display()
    {
        Console.WriteLine(GetInfo());
    }


    public virtual void CalculateBonus()
    {
        double bonus = Salary * 0.1;
        Console.WriteLine($"Бонус сотрудника {Name}: {bonus} рублей");
    }

    public int GetWorkExperience()
    {
        return DateTime.Now.Year - HireDate.Year;
    }

    public virtual void AttendMeeting()
    {
        Console.WriteLine($"Сотрудник {Name} присутствует на собрании");
    }
}


public class Manager : Employee
{
    public int SubordinatesCount;
    public string ManagementLevel; 
    public double Budget; 

    public Manager(string name, int age, int salary, string department, int employeeId, DateTime hireDate, 
                  int subordinatescount, string managementLevel, double budget) 
                  : base(name, age, salary, department, employeeId, hireDate)
    {
        SubordinatesCount = subordinatescount;
        ManagementLevel = managementLevel;
        Budget = budget;
    }

    public override void Work()
    {
        Console.WriteLine($"Менеджер {Name} управляет командой из {SubordinatesCount} человек");
    }

    public override void Display()
    {
        Console.WriteLine($"{GetInfo()}, Количество подчиненных: {SubordinatesCount}, Уровень управления: {ManagementLevel}");
    }


    public void ConductInterview()
    {
        Console.WriteLine($"Менеджер {Name} проводит собеседование");
    }

    public override void CalculateBonus()
    {
        double bonus = Salary * 0.15 + SubordinatesCount * 1000;
        Console.WriteLine($"Бонус менеджера {Name}: {bonus} рублей");
    }

    public void ApproveBudget()
    {
        Console.WriteLine($"Менеджер {Name} утверждает бюджет в размере {Budget} рублей");
    }
}


public class Developer : Employee
{
    public string Specialization;
    public string ProgrammingLanguage; 
    public int CompletedProjects; 

    public Developer(string name, int age, int salary, string department, int employeeId, DateTime hireDate, 
                    string specialization, string programmingLanguage, int completedProjects) 
                    : base(name, age, salary, department, employeeId, hireDate)
    {
        Specialization = specialization;
        ProgrammingLanguage = programmingLanguage;
        CompletedProjects = completedProjects;
    }

    public override void TakeVacation()
    {
        Console.WriteLine($"Разработчик {Name} взял отпуск, необходимо согласовать.");
    }

    public override void Display()
    {
        Console.WriteLine($"{GetInfo()}, Специализация: {Specialization}, Язык программирования: {ProgrammingLanguage}");
    }

    
    public void WriteCode()
    {
        Console.WriteLine($"Разработчик {Name} пишет код на {ProgrammingLanguage}");
    }

    public override void CalculateBonus()
    {
        double bonus = Salary * 0.12 + CompletedProjects * 5000;
        Console.WriteLine($"Бонус разработчика {Name}: {bonus} рублей");
    }

    public void DebugCode()
    {
        Console.WriteLine($"Разработчик {Name} отлаживает код");
    }
}


public class Designer : Employee
{
    public string DesignStyle;
    public string DesignSoftware; 
    public string PortfolioLink; 

    public Designer(string name, int age, int salary, string department, int employeeId, DateTime hireDate, 
                   string designstyle, string designSoftware, string portfolioLink) 
                   : base(name, age, salary, department, employeeId, hireDate)
    {
        DesignStyle = designstyle;
        DesignSoftware = designSoftware;
        PortfolioLink = portfolioLink;
    }

    public override void Work()
    {
        Console.WriteLine($"Дизайнер {Name} работает над дизайн-проектом в стиле {DesignStyle}");
    }

    public override void Display()
    {
        Console.WriteLine($"{GetInfo()}, Стиль дизайна: {DesignStyle}, Программа: {DesignSoftware}");
    }

    
    public void CreateMockup()
    {
        Console.WriteLine($"Дизайнер {Name} создает макет в {DesignSoftware}");
    }

    public override void CalculateBonus()
    {
        double bonus = Salary * 0.08;
        Console.WriteLine($"Бонус дизайнера {Name}: {bonus} рублей");
    }

    public void PresentDesign()
    {
        Console.WriteLine($"Дизайнер {Name} представляет дизайн-проект");
    }
}


public class SeniorDeveloper : Developer
{
    public int TeamSize;
    public bool IsTechLead;
    public string[] Certifications; 

    public SeniorDeveloper(string name, int age, int salary, string department, int employeeId, DateTime hireDate,
                         string specialization, string programmingLanguage, int completedProjects,
                         int teamSize, bool isTechLead, string[] certifications)
                         : base(name, age, salary, department, employeeId, hireDate, specialization, programmingLanguage, completedProjects)
    {
        TeamSize = teamSize;
        IsTechLead = isTechLead;
        Certifications = certifications;
    }

    public override void Work()
    {
        Console.WriteLine($"Старший разработчик {Name} руководит командой из {TeamSize} человек");
    }

    public override void Display()
    {
        string techLeadStatus = IsTechLead ? "Технический лидер" : "Старший разработчик";
        Console.WriteLine($"{GetInfo()}, {techLeadStatus}, Размер команды: {TeamSize}");
    }

    
    public void MentorJuniorDevelopers()
    {
        Console.WriteLine($"Старший разработчик {Name} обучает младших разработчиков");
    }

    public void ConductCodeReview()
    {
        Console.WriteLine($"Старший разработчик {Name} проводит ревью кода");
    }

    public void DisplayCertifications()
    {
        Console.WriteLine($"Сертификации {Name}: {string.Join(", ", Certifications)}");
    }
}


public interface IProjectManagement
{
    void CreateProject(string projectName);
    void AssignTasks();
    int ManagedProjectsCount { get; set; }
}

public interface ITraining
{
    void ConductTraining(string topic);
    void EvaluateEmployee(Employee employee);
}


public class TeamLead : Developer, IProjectManagement, ITraining
{
    public int ManagedProjectsCount { get; set; }
    public string[] CurrentProjects; 

    public TeamLead(string name, int age, int salary, string department, int employeeId, DateTime hireDate,
                   string specialization, string programmingLanguage, int completedProjects,
                   int managedProjectsCount, string[] currentProjects)
                   : base(name, age, salary, department, employeeId, hireDate, specialization, programmingLanguage, completedProjects)
    {
        ManagedProjectsCount = managedProjectsCount;
        CurrentProjects = currentProjects;
    }

    
    public void CreateProject(string projectName)
    {
        Console.WriteLine($"Тимлид {Name} создает проект: {projectName}");
        ManagedProjectsCount++;
    }

    public void AssignTasks()
    {
        Console.WriteLine($"Тимлид {Name} распределяет задачи среди команды");
    }

    
    public void ConductTraining(string topic)
    {
        Console.WriteLine($"Тимлид {Name} проводит обучение по теме: {topic}");
    }

    public void EvaluateEmployee(Employee employee)
    {
        Console.WriteLine($"Тимлид {Name} оценивает работу сотрудника {employee.Name}");
    }

    public override void Display()
    {
        Console.WriteLine($"{GetInfo()}, Управляет проектами: {ManagedProjectsCount}, Текущие проекты: {string.Join(", ", CurrentProjects)}");
    }

    
    public void CoordinateTeam()
    {
        Console.WriteLine($"Тимлид {Name} координирует работу команды");
    }
}


        Console.WriteLine("--- Простое наследование ---");
        Employee manager = new Manager("Элиз Бентли", 39, 118000, "Маркетинг", 1001, new DateTime(2018, 5, 15), 
                                     23, "Старший менеджер", 5000000);
        Employee developer = new Developer("Томас Мюллер", 37, 95000, "IT", 1002, new DateTime(2020, 3, 10), 
                                         "Веб-разработка", "C#", 12);
        Employee designer = new Designer("Кэтти Дейва", 29, 89000, "Дизайн", 1003, new DateTime(2021, 8, 22), 
                                       "Модерн", "Figma", "portfolio.com/katty");

        manager.Display(); manager.Work(); manager.CalculateBonus();
        Console.WriteLine($"Стаж работы: {manager.GetWorkExperience()} лет");
        ((Manager)manager).ApproveBudget();
        
        Console.WriteLine();
        developer.Display(); developer.TakeVacation(); developer.CalculateBonus();
        ((Developer)developer).WriteCode();
        
        Console.WriteLine();
        designer.Display(); designer.Work(); designer.CalculateBonus();
        ((Designer)designer).CreateMockup();

        Console.WriteLine("--- Сложное наследование ---");
        SeniorDeveloper seniorDev = new SeniorDeveloper("Алексей Петров", 35, 150000, "IT", 1004, new DateTime(2017, 6, 20),
                                                      "Backend разработка", "Java", 25, 5, true, 
                                                      new string[] { "AWS Certified", "Oracle Java Certified" });
        seniorDev.Display();
        seniorDev.Work();
        seniorDev.MentorJuniorDevelopers();
        seniorDev.DisplayCertifications();
        Console.WriteLine($"Стаж работы: {seniorDev.GetWorkExperience()} лет");

        Console.WriteLine("--- Множественное наследование ---");
        TeamLead teamLead = new TeamLead("Мария Смирнова", 42, 180000, "IT", 1005, new DateTime(2015, 2, 14),
                                       "Full-stack разработка", "JavaScript", 40, 8,
                                       new string[] { "Web Portal", "Mobile App" });
        teamLead.Display();
        teamLead.Work();
        teamLead.CreateProject("Новый CRM");
        teamLead.AssignTasks();
        teamLead.ConductTraining("Микросервисная архитектура");
        teamLead.EvaluateEmployee(developer);
        teamLead.CoordinateTeam();
        Console.WriteLine($"Стаж работы: {teamLead.GetWorkExperience()} лет");
    
```
