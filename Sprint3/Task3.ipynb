```
public class Employee
{
    public string Name;
    public int Age;
    public int Salary;
    public string Department;
    public DateTime HireDate;
    public int VacationDays;
    public string Email;

    public Employee(string name, int age, int salary, string department, DateTime hireDate, int vacationDays, string email)
    {
        Name = name;
        Age = age;
        Salary = salary;
        Department = department;
        HireDate = hireDate;
        VacationDays = vacationDays;
        Email = email;
    }

    public Employee(string name, int age, int salary, string department)
        : this(name, age, salary, department, DateTime.Now, 28, $"{name.ToLower().Replace(" ", ".")}@company.com")
    {
    }

    public Employee(string name, int age, int salary)
        : this(name, age, salary, "General", DateTime.Now, 28, $"{name.ToLower().Replace(" ", ".")}@company.com")
    {
    }

    public virtual string GetInfo()
    {
        return $"\nИмя: {Name}, \nВозраст: {Age}, \nЗарплата: {Salary} рублей, \nОтдел: {Department}";
    }

    public virtual void Work()
    {
        Console.WriteLine($"Сотрудник {Name} работает");
    }

    public virtual void Work(string task)
    {
        Console.WriteLine($"Сотрудник {Name} работает над задачей: {task}");
    }

    public virtual void TakeVacation()
    {
        Console.WriteLine($"Сотрудник {Name} взял отпуск");
    }

    public virtual void TakeVacation(int days)
    {
        if (days <= VacationDays)
        {
            VacationDays -= days;
            Console.WriteLine($"Сотрудник {Name} взял отпуск на {days} дней. Осталось дней: {VacationDays}");
        }
        else
        {
            Console.WriteLine($"Недостаточно дней отпуска. Доступно: {VacationDays}");
        }
    }

    public virtual void Display()
    {
        Console.WriteLine(GetInfo());
    }

    public virtual void CalculateBonus()
    {
        double bonus = Salary * 0.1;
        Console.WriteLine($"Бонус сотрудника {Name}: {bonus} рублей");
    }

    public virtual void SendEmail(string message)
    {
        Console.WriteLine($"Отправлено email на {Email}: {message}");
    }

    public virtual int CalculateExperience()
    {
        return (DateTime.Now - HireDate).Days / 365;
    }

    public override string ToString()
    {
        return $"Employee: {Name}, {Age} лет, {Department}";
    }
}

public class Manager : Employee
{
    public int SubordinatesCount;
    public string ManagementLevel;
    public double Budget;

    public Manager(string name, int age, int salary, int subordinatesCount, string managementLevel, double budget, string department, DateTime hireDate, int vacationDays, string email) 
        : base(name, age, salary, department, hireDate, vacationDays, email)
    {
        SubordinatesCount = subordinatesCount;
        ManagementLevel = managementLevel;
        Budget = budget;
    }

    public Manager(string name, int age, int salary, int subordinatesCount, string department)
        : base(name, age, salary, department)
    {
        SubordinatesCount = subordinatesCount;
        ManagementLevel = "Middle";
        Budget = 500000;
    }

    public Manager(string name, int age, int salary, int subordinatesCount)
        : this(name, age, salary, subordinatesCount, "General")
    {
    }

    public override void Work()
    {
        Console.WriteLine($"Менеджер {Name} управляет командой из {SubordinatesCount} человек");
    }

    public override void Work(string task)
    {
        Console.WriteLine($"Менеджер {Name} управляет выполнением задачи: {task}");
    }

    public override void Display()
    {
        Console.WriteLine($"{GetInfo()}, Количество подчиненных: {SubordinatesCount}, \nУровень управления: {ManagementLevel}, \nБюджет: {Budget}");
    }

    public override void CalculateBonus()
    {
        double bonus = Salary * 0.15 + SubordinatesCount * 1000;
        Console.WriteLine($"Бонус менеджера {Name}: {bonus} рублей");
    }

    public void ConductMeeting()
    {
        Console.WriteLine($"Менеджер {Name} проводит собрание с {SubordinatesCount} подчиненными");
    }

    public override string ToString()
    {
        return $"Manager: {Name}, \nУровень: {ManagementLevel}, \nПодчиненные: {SubordinatesCount}";
    }
}

public class Developer : Employee
{
    public string Specialization;
    public string ProgrammingLanguage;
    public int CompletedProjects;

    public Developer(string name, int age, int salary, string specialization, string programmingLanguage, int completedProjects, string department, DateTime hireDate, int vacationDays, string email) 
        : base(name, age, salary, department, hireDate, vacationDays, email)
    {
        Specialization = specialization;
        ProgrammingLanguage = programmingLanguage;
        CompletedProjects = completedProjects;
    }

    public Developer(string name, int age, int salary, string specialization, string department)
        : base(name, age, salary, department)
    {
        Specialization = specialization;
        ProgrammingLanguage = "C#";
        CompletedProjects = 0;
    }

    public Developer(string name, int age, int salary, string specialization)
        : this(name, age, salary, specialization, "IT")
    {
    }

    public override void TakeVacation()
    {
        Console.WriteLine($"Разработчик {Name} взял отпуск, необходимо согласовать.");
    }

    public override void TakeVacation(int days)
    {
        if (days <= 14)
        {
            base.TakeVacation(days);
        }
        else
        {
            Console.WriteLine($"Разработчик {Name} не может взять отпуск более 14 дней подряд");
        }
    }

    public override void Display()
    {
        Console.WriteLine($"{GetInfo()}, \nСпециализация: {Specialization}, \nЯзык программирования: {ProgrammingLanguage}, \nЗавершенные проекты: {CompletedProjects}");
    }

    public override void CalculateBonus()
    {
        double bonus = Salary * 0.12 + CompletedProjects * 5000;
        Console.WriteLine($"Бонус разработчика {Name}: {bonus} рублей");
    }

    public void WriteCode()
    {
        Console.WriteLine($"Разработчик {Name} пишет код на {ProgrammingLanguage}");
    }

    public override string ToString()
    {
        return $"\nDeveloper: {Name}, \nСпециализация: {Specialization}, \nЯзык: {ProgrammingLanguage}";
    }
}

public class Designer : Employee
{
    public string DesignStyle;
    public string DesignSoftware;
    public int DesignsCreated;

    public Designer(string name, int age, int salary, string designStyle, string designSoftware, int designsCreated, string department, DateTime hireDate, int vacationDays, string email) 
        : base(name, age, salary, department, hireDate, vacationDays, email)
    {
        DesignStyle = designStyle;
        DesignSoftware = designSoftware;
        DesignsCreated = designsCreated;
    }

    public Designer(string name, int age, int salary, string designStyle, string department)
        : base(name, age, salary, department)
    {
        DesignStyle = designStyle;
        DesignSoftware = "Adobe Creative Suite";
        DesignsCreated = 0;
    }

    public Designer(string name, int age, int salary, string designStyle)
        : this(name, age, salary, designStyle, "Design")
    {
    }

    public override void Work()
    {
        Console.WriteLine($"Дизайнер {Name} работает над дизайн-проектом в стиле {DesignStyle}");
    }

    public override void Work(string task)
    {
        Console.WriteLine($"Дизайнер {Name} создает дизайн для: {task} используя {DesignSoftware}");
    }

    public override void Display()
    {
        Console.WriteLine($"{GetInfo()}, \nСтиль дизайна: {DesignStyle}, \nПрограмма: {DesignSoftware}, \nСоздано дизайнов: {DesignsCreated}");
    }

    public override void CalculateBonus()
    {
        double bonus = Salary * 0.1 + DesignsCreated * 3000;
        Console.WriteLine($"Бонус дизайнера {Name}: {bonus} рублей");
    }

    public void CreateDesign()
    {
        DesignsCreated++;
        Console.WriteLine($"Дизайнер {Name} создал новый дизайн. \nВсего создано: {DesignsCreated}");
    }

    public override string ToString()
    {
        return $"Designer: {Name}, \nСтиль: {DesignStyle}, \nПрограмма: {DesignSoftware}";
    }
}

public class EmployeeManager<T> where T : Employee
{
    private List<T> employees = new List<T>();

    public void AddEmployee(T employee)
    {
        employees.Add(employee);
        Console.WriteLine($"Сотрудник {employee.Name} добавлен в менеджер");
    }

    public void RemoveEmployee(T employee)
    {
        if (employees.Remove(employee))
        {
            Console.WriteLine($"Сотрудник {employee.Name} удален из менеджера");
        }
    }

    public void DisplayAllEmployees()
    {
        Console.WriteLine($"\nВсе сотрудники в менеджере ({typeof(T).Name}):");
        foreach (var employee in employees)
        {
            employee.Display();
        }
    }

    public void WorkAllEmployees()
    {
        Console.WriteLine($"\nРабота всех сотрудников в менеджере ({typeof(T).Name}):");
        foreach (var employee in employees)
        {
            employee.Work();
        }
    }

    public T FindEmployeeByName(string name)
    {
        return employees.Find(e => e.Name.Equals(name, StringComparison.OrdinalIgnoreCase));
    }

    public List<T> GetEmployees()
    {
        return new List<T>(employees);
    }
}

public class SalaryReport<T> where T : Employee
{
    public void GenerateReport(List<T> employees)
    {
        if (employees.Count == 0)
        {
            Console.WriteLine("Нет сотрудников для отчета");
            return;
        }

        Console.WriteLine($"\nОтчет по зарплатам ({typeof(T).Name}):");
        double totalSalary = 0;
        
        foreach (var employee in employees)
        {
            Console.WriteLine($"{employee.Name}: {employee.Salary} рублей");
            totalSalary += employee.Salary;
        }
        
        Console.WriteLine($"\nОбщая сумма зарплат: {totalSalary} рублей");
        Console.WriteLine($"\nСредняя зарплата: {totalSalary / employees.Count} рублей");
    }
}
        Employee first = new Manager("Элиз Бентли", 39, 118000, 23);
        Employee second = new Developer("Томас Мюллер", 37, 95000, "Веб-разработка");
        Employee third = new Designer("Кэтти Дейва", 29, 89000, "Модерн");

   
        Console.WriteLine("\nДемонстрация полиморфизма");
        first.Display(); first.Work(); first.Work("новый проект");
        second.Display(); second.TakeVacation(); second.TakeVacation(10);
        third.Display(); third.Work(); third.Work("логотип компании");

        Console.WriteLine("\nДемонстрация переопределенных методов");
        first.CalculateBonus();
        second.CalculateBonus();
        third.CalculateBonus();

        Console.WriteLine("\nДемонстрация новых методов");
        ((Manager)first).ConductMeeting();
        ((Developer)second).WriteCode();
        ((Designer)third).CreateDesign();

        Console.WriteLine("\nДемонстрация переопределенного ToString");
        Console.WriteLine(first.ToString());
        Console.WriteLine(second.ToString());
        Console.WriteLine(third.ToString());

        Console.WriteLine("\nДемонстрация Generic классов");
        
        var managerManager = new EmployeeManager<Manager>();
        var developerManager = new EmployeeManager<Developer>();
        var designerManager = new EmployeeManager<Designer>();
        
        managerManager.AddEmployee((Manager)first);
        developerManager.AddEmployee((Developer)second);
        designerManager.AddEmployee((Designer)third);
        
        Manager newManager = new Manager("Анна Смит", 42, 125000, 15, "HR");
        Developer newDeveloper = new Developer("Джон Доу", 32, 110000, "Бэкенд", "IT");
        Designer newDesigner = new Designer("Мария Иванова", 35, 92000, "Минимализм", "Design");
        
        managerManager.AddEmployee(newManager);
        developerManager.AddEmployee(newDeveloper);
        designerManager.AddEmployee(newDesigner);
        
        managerManager.DisplayAllEmployees();
        developerManager.DisplayAllEmployees();
        designerManager.DisplayAllEmployees();
        
        managerManager.WorkAllEmployees();
        developerManager.WorkAllEmployees();
        designerManager.WorkAllEmployees();

        var salaryReport = new SalaryReport<Employee>();
        var allEmployees = new List<Employee> { first, second, third, newManager, newDeveloper, newDesigner };
        salaryReport.GenerateReport(allEmployees);

        Console.WriteLine("\nПоиск сотрудника:");
        var foundEmployee = developerManager.FindEmployeeByName("Томас Мюллер");
        if (foundEmployee != null)
        {
            Console.WriteLine("Найден сотрудник:");
            foundEmployee.Display();
        }
```
